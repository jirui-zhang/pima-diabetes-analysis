```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Pima Indian Diabetes dataset
data = pd.read_csv('C:/Users/Membo/Downloads//pima.csv')

# Set the style for the plots
sns.set(style="whitegrid")

# Bar Chart: Comparison of Average Glucose Levels by Diabetes Outcome
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x="Outcome", y="Glucose", palette=["green", "red"])
plt.title("Comparison of Average Glucose Levels by Diabetes Outcome")
plt.xlabel("Diabetes Outcome (0 = Non-Diabetic, 1 = Diabetic)")
plt.ylabel("Average Glucose Level")
plt.show()


```

```python
# Pie Chart: Proportion of Diabetic and Non-Diabetic Cases
outcome_counts = data['Outcome'].value_counts()
labels = ['Non-Diabetic', 'Diabetic']
colors = ['#66c2a5', '#fc8d62']

plt.figure(figsize=(5, 5))
plt.pie(outcome_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Proportion of Diabetic and Non-Diabetic Cases")
plt.show()
```

```python
# Calculate the correlation matrix
corr = data.corr()

# Create the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Pima Diabetes Features")
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=data, x="Age", y="Glucose", hue="Outcome", palette="husl")
plt.title("Age vs. Glucose by Diabetes Outcome")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.legend(title="Diabetes Outcome", labels=["Non-Diabetic", "Diabetic"])
plt.show()
```

```python
# Create the histogram of glucose levels
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x="Glucose", hue="Outcome", kde=True, bins=30, palette="Set2")
plt.title("Glucose Level Distribution by Diabetes Outcome")
plt.xlabel("Glucose Levels")
plt.ylabel("Count of Individuals")
plt.legend(title="Diabetes Outcome", labels=["Non-Diabetic", "Diabetic"])
plt.show()
```

```python
# Create the box plot of BMI grouped by Outcome
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x="Outcome", y="BMI", palette="Set1")
plt.title("BMI Distribution by Diabetes Outcome")
plt.xlabel("Diabetes Outcome (0 = Non-Diabetic, 1 = Diabetic)")
plt.ylabel("BMI")
plt.xticks([0, 1], ["Non-Diabetic", "Diabetic"])  # Custom x-axis labels
plt.show()
```
